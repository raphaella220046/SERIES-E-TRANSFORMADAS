```python
import numpy as np
import matplotlib.pyplot as plt

def fourier_series(x, ck, k_values):
    # Calcula a soma da série de Fourier para um conjunto de coeficientes ck e valores de k
    series_sum = np.zeros_like(x, dtype=np.complex128)
    for ck, k in zip(ck, k_values):
        series_sum += ck * np.exp(1j * k * x)
    return series_sum

# Parâmetros da função
x = np.linspace(-np.pi, np.pi, 1000)  # Intervalo de x
k_values = np.arange(-10, 11)  # Valores de k para a soma da série de Fourier
ck = np.random.rand(len(k_values)) + 1j * np.random.rand(len(k_values))  # Coeficientes complexos aleatórios

# Calcula a série de Fourier
series = fourier_series(x, ck, k_values)

# Plota a parte real e imaginária da série de Fourier
plt.plot(x, np.real(series), label='Real')
plt.plot(x, np.imag(series), label='Imaginário')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Série de Fourier')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def fourier_series(x, ck, k_values):
    # Calcula a soma da série de Fourier para um conjunto de coeficientes ck e valores de k
    series_sum = np.zeros_like(x, dtype=np.complex128)
    for ck, k in zip(ck, k_values):
        series_sum += ck * np.exp(1j * k * x)
    return series_sum

# Parâmetros da função
x = np.linspace(-np.pi, np.pi, 1000)  # Intervalo de x
k_values = np.arange(-10, 11)  # Valores de k para a soma da série de Fourier
ck = np.random.rand(len(k_values)) + 1j * np.random.rand(len(k_values))  # Coeficientes complexos aleatórios

# Calcula a série de Fourier
series = fourier_series(x, ck, k_values)

# Plota o módulo e a fase da série de Fourier
plt.figure(figsize=(10, 4))

# Plot do módulo (amplitude)
plt.subplot(1, 2, 1)
plt.plot(x, np.abs(series))
plt.xlabel('x')
plt.ylabel('Amplitude')
plt.title('Módulo')

# Plot da fase
plt.subplot(1, 2, 2)
plt.plot(x, np.angle(series))
plt.xlabel('x')
plt.ylabel('Fase')
plt.title('Fase')

plt.tight_layout()
plt.show()
```
